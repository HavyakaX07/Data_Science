## Cleaning

```python
#Removing Extra space
df['Salary']=df['Salary'].str.strip()
```

```python
df.replace({'Salary':{'<=50K':0,'>50K':1}},inplace=True)
```

```python
df=df.rename(columns={'Salary':'Salary_greaterthan50K'})
```

```python
#Removing all white spaces for string var
for cl,dt in zip(df.columns,df.dtypes):
    if dt==object:
        df[cl]=df[cl].str.strip()
```

```python
df.replace({'Workclass':{'?':'Unknown'},'Occupation':{'?':'Unknown'}},inplace=True)
df['Workclass'].value_counts()
```

```python
df.drop(['Country','Race','Relationship'],axis=1,inplace=True)
```
