```python
#Finding missing values
df.isnull().sum()
```

```python
#How many columns have missing values greater than 0
[feature for feature in df.columns if df[feature].isnull().sum()>0]
```

```python
#Country code and merging
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,left_on='Country Code',right_on='Country Code',how='inner')
final_df.head()
```

```python
final_df['Country'].value_counts()
```

```python
CountryList=final_df['Country'].value_counts().index
CountryValues=final_df['Country'].value_counts().values
```

```python
#Creating pie charts
plt.pie(CountryValues[:3],labels=CountryList[:3],autopct='%1.2f%%')
```

Observation: Most of the orders of Zomato was came from India.Second place USA and next UK

```python
#Handling Numerical variables
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations
1. Rating is >4.5 says food is Excellent
2. Rating is 4 to 4.4 says food is Very good
3. Rating is 3.5 to 3.9 say food is Good
4. Rating is 2.5 to 3.4 say food is Average
5. Ratind 1.8 to 2.4 says food is poor
6. Rating 0 says food is not rated and indicated in white color

```python
import matplotlib as mt
mt.rcParams['figure.figsize']=(12,6) #rcParams is used control and modify some of the properties of plot like width and color
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

## Observation
1. Not rated count is very high 
2. Maximum ratings is between 2.5 to 4

```python
mt.rcParams['figure.facecolor']='white'
sns.barplot(x='Rating color',y='Rating Count',data=ratings,palette=['gray','red','orange','yellow','green','green'])
```
